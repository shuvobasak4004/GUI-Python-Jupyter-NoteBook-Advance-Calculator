```python
import math
import tkinter as tk

def button_click(number):
    current = entry.get()
    entry.delete(0, tk.END)
    entry.insert(tk.END, current + str(number))

def button_clear():
    entry.delete(0, tk.END)

def button_add():
    first_number = entry.get()
    global f_num
    global math_operation
    math_operation = "addition"
    f_num = float(first_number)
    entry.delete(0, tk.END)

def button_subtract():
    first_number = entry.get()
    global f_num
    global math_operation
    math_operation = "subtraction"
    f_num = float(first_number)
    entry.delete(0, tk.END)

def button_multiply():
    first_number = entry.get()
    global f_num
    global math_operation
    math_operation = "multiplication"
    f_num = float(first_number)
    entry.delete(0, tk.END)

def button_divide():
    first_number = entry.get()
    global f_num
    global math_operation
    math_operation = "division"
    f_num = float(first_number)
    entry.delete(0, tk.END)

def button_equal():
    second_number = entry.get()
    entry.delete(0, tk.END)

    if math_operation == "addition":
        result = f_num + float(second_number)
    elif math_operation == "subtraction":
        result = f_num - float(second_number)
    elif math_operation == "multiplication":
        result = f_num * float(second_number)
    elif math_operation == "division":
        if float(second_number) != 0:
            result = f_num / float(second_number)
        else:
            result = "Error: Division by zero"
    else:
        result = ""

    entry.insert(tk.END, result)

def button_sin():
    num = float(entry.get())
    result = math.sin(num)
    entry.delete(0, tk.END)
    entry.insert(tk.END, result)

def button_cos():
    num = float(entry.get())
    result = math.cos(num)
    entry.delete(0, tk.END)
    entry.insert(tk.END, result)

def button_tan():
    num = float(entry.get())
    result = math.tan(num)
    entry.delete(0, tk.END)
    entry.insert(tk.END, result)

def button_log():
    num = float(entry.get())
    result = math.log10(num)
    entry.delete(0, tk.END)
    entry.insert(tk.END, result)

def button_exp():
    num = float(entry.get())
    result = math.exp(num)
    entry.delete(0, tk.END)
    entry.insert(tk.END, result)

def button_ln():
    num = float(entry.get())
    result = math.log(num)
    entry.delete(0, tk.END)
    entry.insert(tk.END, result)

def button_sqrt():
    num = float(entry.get())
    result = math.sqrt(num)
    entry.delete(0, tk.END)
    entry.insert(tk.END, result)

def button_factorial():
    num = int(entry.get())
    result = math.factorial(num)
    entry.delete(0, tk.END)
    entry.insert(tk.END, result)

def button_abs():
    num = float(entry.get())
    result = abs(num)
    entry.delete(0, tk.END)
    entry.insert(tk.END, result)

def button_pi():
    entry.insert(tk.END, math.pi)

# Create the main window
window = tk.Tk()
window.title("Scientific Calculator")
window.configure(bg="hotpink")

# Create the entry field
entry = tk.Entry(window, width=35, borderwidth=5)
entry.grid(row=0, column=0, columnspan=5, padx=10, pady=10)

# Create the number buttons
button_1 = tk.Button(window, text="1", padx=20, pady=10, bg="lightblue", command=lambda: button_click(1))
button_2 = tk.Button(window, text="2", padx=20, pady=10, bg="lightblue", command=lambda: button_click(2))
button_3 = tk.Button(window, text="3", padx=20, pady=10, bg="lightblue", command=lambda: button_click(3))
button_4 = tk.Button(window, text="4", padx=20, pady=10, bg="lightblue", command=lambda: button_click(4))
button_5 = tk.Button(window, text="5", padx=20, pady=10, bg="lightblue", command=lambda: button_click(5))
button_6 = tk.Button(window, text="6", padx=20, pady=10, bg="lightblue", command=lambda: button_click(6))
button_7 = tk.Button(window, text="7", padx=20, pady=10, bg="lightblue", command=lambda: button_click(7))
button_8 = tk.Button(window, text="8", padx=20, pady=10, bg="lightblue", command=lambda: button_click(8))
button_9 = tk.Button(window, text="9", padx=20, pady=10, bg="lightblue", command=lambda: button_click(9))
button_0 = tk.Button(window, text="0", padx=20, pady=10, bg="lightblue", command=lambda: button_click(0))

# Create the operator buttons
button_add = tk.Button(window, text="+", padx=19, pady=10, bg="lightgreen", command=button_add)
button_subtract = tk.Button(window, text="-", padx=20, pady=10, bg="lightgreen", command=button_subtract)
button_multiply = tk.Button(window, text="*", padx=20, pady=10, bg="lightgreen", command=button_multiply)
button_divide = tk.Button(window, text="/", padx=20, pady=10, bg="lightgreen", command=button_divide)
button_equal = tk.Button(window, text="=", padx=45, pady=10, bg="lightgreen", command=button_equal)
button_clear = tk.Button(window, text="Clear", padx=80, pady=10, bg="lightcoral", command=button_clear)

# Create the scientific buttons
button_sin = tk.Button(window, text="sin", padx=16, pady=10, bg="lightyellow", command=button_sin)
button_cos = tk.Button(window, text="cos", padx=14, pady=10, bg="lightyellow", command=button_cos)
button_tan = tk.Button(window, text="tan", padx=14, pady=10, bg="lightyellow", command=button_tan)
button_log = tk.Button(window, text="log", padx=15, pady=10, bg="lightyellow", command=button_log)
button_exp = tk.Button(window, text="exp", padx=14, pady=10, bg="lightyellow", command=button_exp)
button_ln = tk.Button(window, text="ln", padx=19, pady=10, bg="lightyellow", command=button_ln)
button_sqrt = tk.Button(window, text="sqrt", padx=12, pady=10, bg="lightyellow", command=button_sqrt)
button_factorial = tk.Button(window, text="factorial", padx=4, pady=10, bg="lightyellow", command=button_factorial)
button_abs = tk.Button(window, text="abs", padx=16, pady=10, bg="lightyellow", command=button_abs)
button_pi = tk.Button(window, text="pi", padx=21, pady=10, bg="lightyellow", command=button_pi)

# Position the buttons on the grid
button_7.grid(row=1, column=0)
button_8.grid(row=1, column=1)
button_9.grid(row=1, column=2)
button_divide.grid(row=1, column=3)
button_clear.grid(row=1, column=4)

button_4.grid(row=2, column=0)
button_5.grid(row=2, column=1)
button_6.grid(row=2, column=2)
button_multiply.grid(row=2, column=3)
button_sin.grid(row=2, column=4)

button_1.grid(row=3, column=0)
button_2.grid(row=3, column=1)
button_3.grid(row=3, column=2)
button_subtract.grid(row=3, column=3)
button_cos.grid(row=3, column=4)

button_0.grid(row=4, column=0)
button_equal.grid(row=4, column=1, columnspan=2)
button_add.grid(row=4, column=3)
button_tan.grid(row=4, column=4)

button_log.grid(row=5, column=0)
button_exp.grid(row=5, column=1)
button_ln.grid(row=5, column=2)
button_sqrt.grid(row=5, column=3)
button_factorial.grid(row=5, column=4)

button_abs.grid(row=6, column=0)
button_pi.grid(row=6, column=1)

# Start the main event loop
window.mainloop()
```
